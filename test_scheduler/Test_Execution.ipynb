```python
from json_parser import JSONParser
from db_handler import DBHandler, logger

DBNAME = "memories.db"

db_handler = DBHandler("memories.db")
file_list = JSONParser.grep_all_json_files_in_directory('config_files/memory_configs')
memory_files = []
for file in file_list:
    memory_files.append(JSONParser.read_json_file('config_files/memory_configs', file))

try:
    # Initialize the database and create the table
    db_handler.initialize()
    db_handler.clear_all_tables()
    db_handler.create_memory_table()
    db_handler.create_memory_instance_table()
    for mem_config in memory_files:
        db_handler.add_memory_type(mem_config)
#    db_handler.get_all_tables()


    # Query and display all data from the `memories` table
    records = db_handler.query_all_memories()
    print("RECORDS", records)
    for record in records:
        print(record)

except Exception as e:
    logger.error(f"An error occurred: {e}")

finally:
    # Ensure the database connection is closed
    db_handler.close()
```

```python
import socket
import os
import json

def create_udp_socket():
    # Create a UDP socket
    udp_socket = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

    # Define the server address and port
    server_address = ("132.231.14.92", 5025)

    try:
        # Check if the config file exists
        config_file_path = 'config_files/config.json'
        if not os.path.exists(config_file_path):
            print(f"Error: Config file '{config_file_path}' does not exist.")
            return

        # Read the config file
        with open(config_file_path, "r") as config_file:
            config = config_file.read()
            print(f"Sending: {config}")

        # Send the message to the server
        udp_socket.sendto(config.encode(), server_address)

        # Receive the response from the server
        response, server = udp_socket.recvfrom(4096)  # Buffer size of 4096 bytes
        out_dict = json.loads(response.decode()[:-1])
        print(f"Received: {out_dict} from {server}")

    except (OSError, socket.error) as e:
        print(f"Socket error: {e}")
    except Exception as e:
        print(f"An error occurred: {e}")
    finally:
        # Close the socket
        udp_socket.close()

if __name__ == "__main__":
    create_udp_socket()


```
